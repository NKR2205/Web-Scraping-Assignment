# Q1. Scrape the details of most viewed videos on YouTube from Wikipedia.

```python
# Connecting to the driver
driver = webdriver.Chrome(r"C:\Users\navee\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
# Opening web Page
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
# Scraping Rank

rank = []

rk = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[1]')
for i in rk:
    rank.append(i.text)

# Scraping Name

Name = []

nam = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[2]')
for i in nam:
    Name.append(i.text)

# Scraping Artist

Artist = []

art = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[3]')
for i in art:
    Artist.append(i.text)
    
# Scrapping Upload Date

date = []

dt = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[5]')
for i in dt:
    date.append(i.text)

# Scrapping views

views = []

vw = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[4]')
for i in vw:
    views.append(i.text)
```

```python
# Removing numerical data from Name

import re

name3 = []
for i in range(len(Name)):
    if i==len(Name)-31:
         break
    else:
        name2 = re.sub(r'\[[0-9]*\]', '', Name[i])
    name3.append(name2)
name3
```

```python
# Creating DataFrame

yt = pd.DataFrame({"Rank":rank, "Name":name3,"Artist":Artist,"Upload Date":date,"Views":views})
yt
```

# Q2.  Scrape the details team India’s international fixtures from bcci.tv.

```python
# Connecting to the driver
driver = webdriver.Chrome(r"C:\Users\navee\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
# Opening web Page
driver.get("http://www.bcci.tv/")

try:
    search1 = driver.find_element(By.XPATH,'//*[@id="navigation"]/ul[1]/li[2]/a')
    search1.click()
except NoSuchElementException:
    search2 = driver.find_element(By.XPATH,'//button[@data-target="#navigation"]')
    search2.click()
except NoSuchElementException:
    search3 = driver.find_element(By.XPATH,'//*[@id="navigation"]/ul[1]/li[2]/a')
    search3.click()
    time.sleep(3)


for _ in range(0,20):
    driver.execute_script("window.scrollBy(0,50)")
search4 = driver.find_element(By.XPATH,'//button[@ng-click="viewMoreMatches()"]')
search4.click()
```

```python
# Scraping Title

title = []

tit = driver.find_elements(By.XPATH,'//span[@class="matchOrderText ng-binding ng-scope"]')
for i in tit:
    title.append(i.text)


# Scraping series

series = []
ser = driver.find_elements(By.XPATH,'//span[@class="ng-binding"]')
for i in ser:
    series.append(i.text)
    

# scraping place
place = []
pl = driver.find_elements(By.XPATH,'//div[@class="fix-place ng-binding ng-scope"]')
for i in pl:
    place.append(i.text)


place3 = []
for i in range(len(place)):
    if i==len(place)-17:
        break
    else:
        place2 = place[i].split(' ')[3:]
    place3.append(place2)

place4 = [' '.join(i) for i in place3]
place4
    
# scraping date
date = []
dt = driver.find_elements(By.XPATH,'//h5[@class="ng-binding"]')
for i in dt:
    date.append(i.text)

    
# scraping time
time = []
tt = driver.find_elements(By.XPATH,'//h5[@class="text-right ng-binding"]')
for i in tt:
    time.append(i.text)
```

```python
# Creating DataFrame

india_fixtures = pd.DataFrame({"Match Title":title,"Series":series,"Place":place4,"Date":date,"Time":time}) 
india_fixtures
```

# Q4. Scrape the details of State-wise GDP of India from statisticstime.com

```python
# Connecting to the driver
driver = webdriver.Chrome(r"C:\Users\navee\Downloads\chromedriver_win32\chromedriver.exe")
time.sleep(5)

# Opening web Page

driver.get("http://statisticstimes.com/")

try:
    search1 = driver.find_element(By.XPATH,'//*[@id="top"]/div[2]/div[2]/button')
    search1.click()
except InvalidSelectorException:
    search1 = driver.find_element(By.XPATH,'//*[@id="top"]/div[2]/div[2]/button')
    search1.click()
except ElementNotInteractableException:
    search1 = driver.find_element(By.XPATH,'//*[@id="top"]/div[2]/div[2]/button')
    search1.click()

try:
    search2 = driver.find_element(By.XPATH,'//*[@id="top"]/div[2]/div[2]/div/a[3]')
    search2.click()
except ElementNotInteractableException:
    search2 = driver.find_element(By.XPATH,'//*[@id="top"]/div[2]/div[2]/div/a[3]')
    search2.click()
    time.sleep(5)

# Ignoring Advertisement

driver.get("https://www.statisticstimes.com/economy/india-statistics.php")
search3 = driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[2]/ul/li[1]/a')
search3.click()
time.sleep(10)


rank_gdp = []
st_gdp = []
GSDP1 = []
GSDP2 = []
share = []
bill = []

rk = driver.find_elements(By.XPATH,'//*[@id="table_id"]/tbody/tr/td[1]')
for i in rk:
    rank_gdp.append(i.text)


st = driver.find_elements(By.XPATH,'//*[@id="table_id"]/tbody/tr/td[2]')
for i in st:
    st_gdp.append(i.text)


gs1 = driver.find_elements(By.XPATH,'//*[@id="table_id"]/tbody/tr/td[4]')
for i in gs1:
    GSDP1.append(i.text)


gs2 = driver.find_elements(By.XPATH,'//*[@id="table_id"]/tbody/tr/td[3]')
for i in gs2:
    GSDP2.append(i.text)


sh = driver.find_elements(By.XPATH,'//*[@id="table_id"]/tbody/tr/td[5]')
for i in sh:
    share.append(i.text)


bl = driver.find_elements(By.XPATH,'//*[@id="table_id"]/tbody/tr/td[6]')
for i in bl:
    bill.append(i.text)
```

```python
# Creating DataFrame

GDP = pd.DataFrame({"Rank":rank_gdp,"State":st_gdp,"GSDP(18-19)":GSDP1," GSDP(19-20)":GSDP2,"Share(18-19)":share,
                    "GDP($ Billion)":bill})
GDP
```
