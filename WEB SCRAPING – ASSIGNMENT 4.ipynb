# Q1. Scrape the details of most viewed videos on YouTube from Wikipedia.

```python
# Connecting to the driver
driver = webdriver.Chrome(r"C:\Users\navee\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
# Opening web Page
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
# Scraping Rank

rank = []

rk = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[1]')
for i in rk:
    rank.append(i.text)

# Scraping Name

Name = []

nam = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[2]')
for i in nam:
    Name.append(i.text)

# Scraping Artist

Artist = []

art = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[3]')
for i in art:
    Artist.append(i.text)
    
# Scrapping Upload Date

date = []

dt = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[5]')
for i in dt:
    date.append(i.text)

# Scrapping views

views = []

vw = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[4]')
for i in vw:
    views.append(i.text)
```

```python
# Removing numerical data from Name

import re

name3 = []
for i in range(len(Name)):
    if i==len(Name)-31:
         break
    else:
        name2 = re.sub(r'\[[0-9]*\]', '', Name[i])
    name3.append(name2)
```

```python
# Creating DataFrame

yt = pd.DataFrame({"Rank":rank, "Name":name3,"Artist":Artist,"Upload Date":date,"Views":views})
yt
```

# Q5. Scrape the details of trending repositories on Github.com

```python
# Connecting to the driver
driver = webdriver.Chrome(r"C:\Users\navee\Downloads\chromedriver_win32\chromedriver.exe")
time.sleep(3)

# Opening web Page
driver.get("https://github.com/")
time.sleep(5)

# Navigating through trending repository
search1 = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')
search1.click()
    
search1 = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/ul[3]/li[3]/a')
search1.click()
time.sleep(3)


# Scraping Repository URL

gt_url = []

url = driver.find_elements(By.XPATH,'/html/body/div[1]/div[4]/main/div[3]/div/div[2]/article/h1/a')
for i in url:
    gt_url.append(i.get_attribute("href"))


# Scraping Name from Title

title = []
for i in range(len(gt_url)):
    if i==len(gt_url)-26:
        break
    else:
        tit = gt_url[i].split('/github.com/')[1:]
    title.append(tit)

    
# Scraping Description

des = []

for url in gt_url[0:25]:
    driver.get(url)
    time.sleep(2)
    
    try:
        desc = driver.find_element(By.XPATH,'//*[@id="repo-content-pjax-container"]/div/div/div[3]/div[2]/div/div[1]/div')
        des.append(desc.text)
    except NoSuchElementException:
        des.append('-')

description = []
for i in range(len(des)):
    if i==len(des)-26:
        break
    else:
        des1 = des[i].split('\n')[1:2]
    description.append(des1)

    
# Scraping Contributors

cont = []

for url in gt_url[0:25]:
    driver.get(url)
    time.sleep(3)
    
    try:
        con = driver.find_element(By.XPATH,"//*[contains(text(),'    Contributors ')]")
        cont.append(con.text)
    except NoSuchElementException:
        cont.append('-')

# Scraping Language

lang = []

for url in gt_url[0:25]:
    driver.get(url)
    time.sleep(3)
    
    try:
        lan = driver.find_elements(By.XPATH,'//ul[@class="list-style-none"]')
        for i in lan[-1:]:
            la = i.text
            lang.append(la)
    except NoSuchElementException:
        lang.append('-')

lang2 = []

for i in range(len(lang)):
    if i == len(lang)-26:
        break
    else:
        lan2 = lang[i].split('\n')
    lang2.append(lan2)
```

```python
# Creating DataFrame

Trending_repository = pd.DataFrame({"Repository Title":title,"Repository Description": description,
                                    "Contributors Count":cont,"Language Used":lang2})
Trending_repository
```

# Q6. Scrape the details of top 100 songs on billboard.com

```python
# Connecting to the driver
driver = webdriver.Chrome(r"C:\Users\navee\Downloads\chromedriver_win32\chromedriver.exe")
time.sleep(2)


# Opening webpage

driver.get("https://www.billboard.com/")
time.sleep(10)

search1 = driver.find_element(By.XPATH,'//*[@id="main-wrapper"]/header/div/div[2]/div/div/div[1]/div[1]/button')
search1.click()
time.sleep(3)

search1 = driver.find_element(By.XPATH,'//*[@id="main-wrapper"]/div[9]/div/div/div/ul/li[1]/ul/li[2]/a')
search1.click()
time.sleep(12)


# Scraping Billboard Hot 100

bill =[]
bi = driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]')
for i in bi:
    bill.append(i.text)

# Scraping song name

song = []
for i in range(len(bill)):
    if i==len(bill)-101:
        break
    else:
        son = bill[i].strip().split(",")[0].strip().split("\n")[0]
    song.append(son)
    
    
# Scraping Artist

artist = []
for i in range(len(bill)):
    if i==len(bill)-101:
        break
    else:
        art = bill[i].strip().split("\n")[1]
    artist.append(art)
    
# Scraping last week rank

lwr = []
for i in range(len(bill)):
    if i==len(bill)-101:
        break
    else:
        lr = bill[i].strip().split("\n")[2]
    lwr.append(lr)
    

# Scraping Peak Rank

pkr = []
for i in range(len(bill)):
    if i==len(bill)-101:
        break
    else:
        pk = bill[i].strip().split("\n")[3]
    pkr.append(pk)
    
    
# Scraping Peak Rank

wb = []
for i in range(len(bill)):
    if i==len(bill)-101:
        break
    else:
        w = bill[i].strip().split("\n")[4]
    wb.append(w)
```

```python
# Creating DataFrame

Billboard_top_100 = pd.DataFrame({"Song Name":song,"Artist":artist,"Last Week Rank":lwr,"Peak Rank":pkr,
                                  "Weeks on Board":wb})
Billboard_top_100
```

```python
# Connecting to the driver
driver = webdriver.Chrome(r"C:\Users\navee\Downloads\chromedriver_win32\chromedriver.exe")
time.sleep(3)

# Opening web page
driver.get("http://www.imdb.com/list/ls095964455/")
time.sleep(10)


title=[]
year=[]
genre=[]
runtime=[]
rating=[]
vote=[]

# Scraping title
tl = driver.find_elements(By.XPATH,'//*[@id="main"]/div/div[3]/div[3]/div/div[2]/h3/a')
for i in tl:
    title.append(i.text)
    
    
# Scraping Year span 
yr = driver.find_elements(By.XPATH,'//span[@class="lister-item-year text-muted unbold"]')
for i in yr:
    year.append(i.text)
    
    
# Scraping genre
gr = driver.find_elements(By.XPATH,'//span[@class="genre"]')
for i in gr:
    genre.append(i.text)

    
# Scraping runtime
rn = driver.find_elements(By.XPATH,'//span[@class="runtime"]')
for i in rn:
    runtime.append(i.text)
    
# Scraping ratings
rt = driver.find_elements(By.XPATH,'//*[@id="main"]/div/div[3]/div[3]/div/div[2]/div[1]/div[1]/span[2]')
for i in rt:
    rating.append(i.text)
    

# Scraping votes
vt = driver.find_elements(By.XPATH,'//span[@name="nv"]')
for i in vt:
    vote.append(i.text)
    

# Creating DataFrame

Tv_series_Most_Watched = pd.DataFrame({"Title":title,"Year":year,"Genre":genre,"Runtime":runtime,
                             "Ratings":rating,"Votes":vote})
Tv_series_Most_Watched
```
